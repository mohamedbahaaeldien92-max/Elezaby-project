```python
import pandas as pd
import numpy as np
import re

def is_garbled(text):
    """
    Detect unreadable / corrupted text.
    """
    if pd.isna(text):
        return True

    text = str(text)

    # نسبة الرموز الغريبة
    weird_ratio = len(re.findall(r'[^\w\s/.-]', text)) / max(len(text), 1)

    # لو فيه نسبة كبيرة من رموز غير طبيعية
    if weird_ratio > 0.3:
        return True

    # لو فيه أنماط معروفة للـ mojibake العربي
    if re.search(r'ط|ظ', text):
        return True

    return False


# تطبيق الاستبدال
df["product_name"] = df["product_name"].apply(
    lambda x: "unknown product name" if is_garbled(x) else x
)

print("Garbled text cleaned successfully ✅")
```

```python
import re
import numpy as np

def clean_product_name(series):
    series = series.astype("string")

    # إزالة المسافات الزائدة
    series = series.str.strip()

    # حذف أي رموز غير حروف/أرقام في بداية النص
    series = series.str.replace(r'^[^a-zA-Z0-9]+', '', regex=True)

    # حذف أي رموز غير حروف/أرقام في نهاية النص
    series = series.str.replace(r'[^a-zA-Z0-9]+$', '', regex=True)

    # استبدال القيم الفارغة
    series = series.replace("", np.nan).fillna("unknown")

    return series


df["product_name"] = clean_product_name(df["product_name"])

print("Product name cleaned successfully ✅")
```
